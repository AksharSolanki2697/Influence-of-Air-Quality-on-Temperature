## Replacing the Null Values with mean of Column

```python

columns = ['PM2.5','PM10','SO2','NO2','CO','O3','TEMP','PRES','DEWP','RAIN','WSPM']

def replace_NaN(x):
    data[x].fillna((data[x].mean()), inplace=True) 

for x in columns:
    replace_NaN(x)
```

## Converting String Direction into Degrees

```python
#Better way to convert String Variables to Degrees

Dictionary = {
    'N' : random.uniform(348.75, 360) or random.uniform(0,11.25), 
    'NNE' : random.uniform(11.25,33.75),
    'NE' : random.uniform(33.75,56.25),
    'ENE': random.uniform(56.25,78.75),
    'E' : random.uniform(78.75,101.25), 
    'ESE' : random.uniform(101.25,123.75), 
    'SE': random.uniform(123.75,146.25),
    'SSE' : random.uniform(146.25,168.75), 
    'S': random.uniform(168.75,191.25), 
    'SSW' : random.uniform(191.25,213.75), 
    'SW' : random.uniform(213.75,236.25),
    'WSW' : random.uniform(236.25,258.75),
    'W' : random.uniform(258.75,281.25), 
    'WNW' : random.uniform(281.25,303.75), 
    'NW' : random.uniform(303.75,326.25),
    'NNW' : random.uniform(326.25,348.75)
             }
```

```python
for i in range(len(data)):
    if data['wd'].iloc[i] in Dictionary.keys():
        data['wd'].iloc[i] = Dictionary[data['wd'].iloc[i]]
    else:
        continue
```

```python
# Unit Conversion from ug/m^3 to ppb or ppm
# SO2 1 ppb = 2.62 µg/m3
# NO2 1 ppb = 1.88 µg/m3
# O3 1 ppb = 2.00 µg/m3
# CO 1 ppb = 1.145 µg/m
def convert_measurements():
    data['SO2'] = data['SO2']/2.62 # Measured in ppb
    data['NO2'] = data['NO2']/1.88  # Measured in ppb 
    data['O3'] = data['O3']/2.00   # Measured in ppb
    data['CO'] = data['CO']/(1.145*1000) #Measured in ppm
    
convert_measurements()
```

```python
# Limiting the values of PM2.5 and PM10 since concentration higher than that will have same AQI. This will reduce  
for i in range(len(data)):
    if data['PM10'].iloc[i] > 605:
        data['PM10'].iloc[i] = 604
    if data['PM2.5'].iloc[i] > 500.4:
        data['PM2.5'].iloc[i] = 500.4
print('PM ranges limited')
```

# Calculating AQI for all Pollutants

## PM2.5

```python
# Function to Calculate AQI of PM2.5 Particulate Matter
def aqi_pm25(a):
    conc_pm = a
#     print(conc_pm)
    aqi_range = [0.0,50.0,100.0,150.0,200.0,300,400,500]
    conc_range = [0,12,35.5,55.5,150.5,250.5,350.5,500.5]
    
    if conc_pm > conc_range[0] and conc_pm <= conc_range[1]:
        return conc_range[0],conc_range[1],aqi_range[0],aqi_range[1]
    
    elif conc_pm > conc_range[1] and conc_pm <= conc_range[2]:
        return conc_range[1],conc_range[2],aqi_range[1],aqi_range[2]

    elif conc_pm > conc_range[2] and conc_pm <= conc_range[3]:
        return conc_range[2],conc_range[3],aqi_range[2],aqi_range[3]

    elif conc_pm > conc_range[3] and conc_pm <= conc_range[4]:
        return conc_range[3],conc_range[4],aqi_range[3],aqi_range[4]

    elif conc_pm > conc_range[4] and conc_pm <= conc_range[5]:
        return conc_range[4],conc_range[5],aqi_range[4],aqi_range[5]

    elif conc_pm > conc_range[5] and conc_pm <= conc_range[6]:
        return conc_range[5],conc_range[6],aqi_range[5],aqi_range[6]
    
    elif conc_pm > conc_range[6] and conc_pm <= conc_range[7]:
        return conc_range[6],conc_range[7],aqi_range[6],aqi_range[7]


def calculate_aqi_pm25(a):
    aqi_current = 0
    conc_current = a
#     print(conc_current)
    conc_low,conc_high,aqi_low,aqi_high = aqi_pm25(conc_current)
#     print(conc_low,conc_high,aqi_low,aqi_high)
    aqi_current = (( aqi_high - aqi_low ) * ( conc_current - conc_low ) / (conc_high - conc_low)) + (aqi_low)
#     print(aqi_current)
    return aqi_current
```

```python
data['PM25_AQI'] = 0
for i in range(len(data)):
    a = data['PM2.5'].iloc[i]
    data['PM25_AQI'].iloc[i] = calculate_aqi_pm25(a)
```
